```python
import pandas as pd
import numpy as np

# No modifications can be made without a selected code snippet.import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Banco de dados.xlsx")
df = df[df['Área colhida'] != '...']
df1 = pd.read_excel("fator de transformação.xlsx")
```

```python
df["Cultura"].unique()
```

```python
df = df[df['Cultura'].isin(df1['Cultura'])]
df
```

```python
# Quantificação utilizando simulação de monte Carlo

resultados = []

for municipio in df['Municípios'].unique():
    df_producao_municipio = df[df['Municípios'] == municipio]

    residuos_simulados = []
    biochar_simulado = []
    sequestro_carbono_simulado = []

    # Iterar sobre as culturas do município atual
    for cultura in df['Cultura']:
        df_coeficientes_cultura = df1[df1['Cultura'] == cultura]
        producao = df[df['Cultura'] == cultura]['Produção'].values[0]

        # Simular a produção com desvio padrão de 20% da média
        producao_simulada = np.random.normal(producao, 0.2 * producao, 10000)
        # Simular o rendimento do biochar (25%) com desvio padrão de 5%
        rendimento_biochar_simulado = np.random.normal(0.25, 0.05 * 0.25, 10000)
        # Simular o conteúdo de carbono do biochar (BCC) com desvio padrão de 10%
        bcc_simulado = np.random.normal(df1['BCC'].mean(), 0.1 * df1['BCC'].mean(), 10000)

        # Iterar sobre os tipos de coeficientes da cultura atual
        for index, coef in df_coeficientes_cultura.iterrows():
            for producao, rendimento_biochar, bcc, area_colhida in zip(producao_simulada, rendimento_biochar_simulado, bcc_simulado, df['Área colhida']):
              
                # Cálculo dos resíduos e do biochar usando RPR e produção simulada
                if not pd.isna(coef['RPR']):
        
                    residuos = coef['RPR'] * coef['CF'] * (1 - coef['MC']) * producao
                    biochar = residuos * rendimento_biochar
                    sequestro_carbono = biochar * bcc

                    residuos_simulados.append(residuos)
                    biochar_simulado.append(biochar)
                    sequestro_carbono_simulado.append(sequestro_carbono)
                    
                # Cálculo dos resíduos e do biochar usando ARP e área colhida
                elif not pd.isna(coef['ARP']):
                
                    residuos = coef['ARP'] * coef['CF'] * (1 - coef['MC']) * area_colhida
                    biochar = residuos * rendimento_biochar
                    sequestro_carbono = biochar * bcc

                
                    residuos_simulados.append(residuos)
                    biochar_simulado.append(biochar)
                    sequestro_carbono_simulado.append(sequestro_carbono)

    # Cálculo dos intervalos de confiança para os resíduos, biochar e sequestro de carbono
    resultados.append({
        'Município': municipio,
        'Soma de Resíduos': np.mean(residuos_simulados),
        'IC Resíduos Inferior': np.percentile(residuos_simulados, 2.5),
        'IC Resíduos Superior': np.percentile(residuos_simulados, 97.5),
        'Biochar Produzido': np.mean(biochar_simulado),
        'IC Biochar Inferior': np.percentile(biochar_simulado, 2.5),
        'IC Biochar Superior': np.percentile(biochar_simulado, 97.5),
        'Sequestro de Carbono (kg)': np.mean(sequestro_carbono_simulado),
        'IC Sequestro Carbono Inferior': np.percentile(sequestro_carbono_simulado, 2.5),
        'IC Sequestro Carbono Superior': np.percentile(sequestro_carbono_simulado, 97.5)
    })


```

```python
df_resultados = pd.DataFrame(resultados)
display(df_resultados)
```
