# 4º Transformação de dados

```python
def soma_linha(coluna):
    return coluna.sum()
# Aplicando a função soma_linha nas linhas do DataFrame
df['Soma'] = df.apply(soma_linha, axis=1)
```

```python
df_valor = df[["Preço Unitário" , "Desconto"]]
df_valor
```

```python
df_valor['Valor Liquido'] = df_valor.apply(lambda linha: linha['Preço Unitário'] - linha['Desconto'], axis=1)
```
