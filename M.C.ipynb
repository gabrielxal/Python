```python
import pandas as pd
import numpy as np
```

```python
dataframe1 = pd.read_excel("Estado - Cultura.xlsx")
dataframe2 = pd.read_excel("Somátorio Cidade.xlsx")
dataframe3 = pd.read_excel("Cidade - Cultura.xlsx")
```

```python
display(dataframe1)
display(dataframe2)
display(dataframe3)
```

```python
rendimento_biochar = 0.25 
n_iter = 10000  
desvio_padrao_rendimento = rendimento_biochar * 0.05  

def monte_carlo_simulation(residuo_biomassa):
    resultados = []
    for _ in range(n_iter):
     
        rendimento_conversao = np.random.normal(rendimento_biochar, desvio_padrao_rendimento)
        quantidade_biochar = residuo_biomassa * rendimento_conversao
        resultados.append(quantidade_biochar)
    
    
    return np.mean(resultados)

def apply_simulation(df):
    df['média biochar (t)'] = df['Resíduos (t)'].apply(monte_carlo_simulation)
    return df
    
df1 = apply_simulation(dataframe1)
df2 = apply_simulation(dataframe2)
df3 = apply_simulation(dataframe3)
```

```python
display(df1)
display(df2)
display(df3)
```

```python
df1.to_excel("Estado - Cultura_montecarlo.xlsx", index=False)
df2.to_excel("Somátorio Cidade_montecarlo.xlsx", index=False)
df3.to_excel("Cidade - Cultura_montecarlo.xlsx", index=False)
```
