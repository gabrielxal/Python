### *Esse código tem objetivo de fazer tratamento dos dados*

```python
import pandas as pd 
import numpy as np
```

```python
# Leitura da planilha
df = pd.read_excel("aevo_-_desafio_estágio_em_sales_ops.xlsx", sheet_name="Dados")
```

```python
# Visualização
df
```

```python
# Verificando se existem duplicatas e remoção de valores ausentes
df.drop_duplicates(inplace=True)
df.dropna()
```

### *Contratos Fechados*

```python
# Filtragem de dados
contratos_fechados = df[df['Resultado da negociação'] == 'Contrato fechado']
contratos_fechados
```

```python
# Estatística
contratos_fechados.describe()
```

```python
# Quantidade de pagamentos
contratos_fechados['Valor do contrato (R$)'].value_counts()
```

```python
# Potencial da sede
contagem_por_localidade = contratos_fechados.groupby('Localidade da sede').size().sort_values(ascending=False)
setor_por_localidade = contratos_fechados.groupby('Localidade da sede')['Setor'].unique()

result = pd.DataFrame({
    'Contagem': contagem_por_localidade,
    'Setor por localidade': setor_por_localidade
})
result.sort_values("Contagem",ascending=False)
```

```python
# Quantas vendas por setores tem nas sedes
vendas_por_sede_setor = contratos_fechados.groupby(['Localidade da sede', 'Setor']).size().reset_index(name='Quantidade de Vendas')
vendas_por_sede_setor = vendas_por_sede_setor.sort_values(by=['Localidade da sede', 'Quantidade de Vendas'], ascending=[True, False])
vendas_por_sede_setor.sort_values("Localidade da sede")
```

### *Contratos Perdidos*

```python
contratos_perdidos = df[df['Resultado da negociação'] == 'Contrato perdido']
perdido_por_sede_setor = contratos_perdidos.groupby(['Localidade da sede', 'Setor']).size().reset_index(name='Quantidade de Vendas')
perdido_por_sede_setor = perdido_por_sede_setor.sort_values(by=['Localidade da sede', 'Quantidade de Vendas'], ascending=[True, False])
perdido_por_sede_setor.sort_values("Localidade da sede")
```

```python
contagem_perdido_localidade = contratos_perdidos.groupby('Localidade da sede').size().sort_values(ascending=False)
setor_perdido_localidade = contratos_perdidos.groupby('Localidade da sede')['Setor'].unique()

result = pd.DataFrame({
    'Contagem': contagem_perdido_localidade,
    'Setor por localidade': setor_perdido_localidade
})
result.sort_values("Contagem",ascending=False)
```

```python
# Contagem de vendas por setor (sem localidade) perdidos
vendas_perdido_setor = contratos_perdidos.groupby('Setor').size().reset_index(name='Quantidade de Perdas')
total_geral = df['Resultado da negociação'].count()
vendas_perdido_setor['Porcentagem Geral (%)'] = (vendas_perdido_setor['Quantidade de Perdas'] / total_geral) * 100
vendas_perdido_setor = vendas_perdido_setor.sort_values(by='Porcentagem Geral (%)', ascending=False)
nova_linha = pd.DataFrame({
    'Setor': ['Taxa de Conversão'],
    'Quantidade de Perdas': [vendas_perdido_setor['Quantidade de Perdas'].sum()],
    'Porcentagem Geral (%)': [vendas_perdido_setor['Quantidade de Perdas'].sum() / total_geral * 100]
})
vendas_perdido_setor = pd.concat([vendas_perdido_setor, nova_linha], ignore_index=True)
vendas_perdido_setor
```

### *Porcentagem - Somente Contratos Fechados*

```python
# Contagem de vendas por setor em cada sede
vendas_por_sede_setor = contratos_fechados.groupby(['Localidade da sede', 'Setor']).size().reset_index(name='Quantidade de Vendas')

total_geral = df['Resultado da negociação'].count()
vendas_por_sede_setor['Porcentagem Geral (%)'] = (vendas_por_sede_setor['Quantidade de Vendas'] / total_geral) * 100
vendas_por_sede_setor = vendas_por_sede_setor.sort_values(by='Porcentagem Geral (%)', ascending=False)
vendas_por_sede_setor.head(10)
```

```python
# Contagem de vendas por setor (sem localidade)
vendas_por_setor = contratos_fechados.groupby('Setor').size().reset_index(name='Quantidade de Vendas')
total_geral = df['Resultado da negociação'].count()
vendas_por_setor['Porcentagem Geral (%)'] = (vendas_por_setor['Quantidade de Vendas'] / total_geral) * 100
vendas_por_setor = vendas_por_setor.sort_values(by='Porcentagem Geral (%)', ascending=False)
nova_linha = pd.DataFrame({
    'Setor': ['Taxa de Conversão'],
    'Quantidade de Vendas': [vendas_por_setor['Quantidade de Vendas'].sum()],
    'Porcentagem Geral (%)': [vendas_por_setor['Quantidade de Vendas'].sum() / total_geral * 100]
})
vendas_por_setor = pd.concat([vendas_por_setor, nova_linha], ignore_index=True)
vendas_por_setor

```

### *Análise Economica*

```python


valor_total_setor = contratos_fechados.groupby('Setor')['Valor do contrato (R$)'].sum().reset_index()
valor_total_setor_localidade = contratos_fechados.groupby(['Setor', 'Localidade da sede'])['Valor do contrato (R$)'].sum().reset_index()
valor_setorT = valor_total_setor.sort_values("Valor do contrato (R$)", ascending=False)
valor_setor_localidadeT = valor_total_setor_localidade.sort_values("Valor do contrato (R$)", ascending=False)
display(valor_setorT)
display(valor_setor_localidadeT.head())
```

```python
# Calcular a média de valor do contrato e tempo de negociação por setor
media_valor_setor = contratos_fechados.groupby('Setor')['Valor do contrato (R$)'].mean().reset_index().sort_values("Valor do contrato (R$)", ascending=False)
media_tempo_negociacao_setor = contratos_fechados.groupby('Setor')['Tempo de negociação (dias)'].mean().reset_index().sort_values("Tempo de negociação (dias)", ascending=False)
display(media_valor_setor)
display(media_tempo_negociacao_setor)
```

### *Tratamento dos Estados*

```python
# Mudança de siglas para nomeclatura
siglas_para_estados = {
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Ceará',
    'DF': 'Distrito Federal',
    'ES': 'Espírito Santo',
    'GO': 'Goiás',
    'MG': 'Minas Gerais',
    'MS': 'Mato Grosso do Sul',
    'MT': 'Mato Grosso',
    'PA': 'Pará',
    'PE': 'Pernambuco',
    'PR': 'Paraná',
    'RJ': 'Rio de Janeiro',
    'RS': 'Rio Grande do Sul',
    'SC': 'Santa Catarina',
    'SP': 'São Paulo'
}

valor_setor_localidadeT['Localidade da sede'] = valor_setor_localidadeT['Localidade da sede'].replace(siglas_para_estados)
vendas_por_sede_setor['Localidade da sede'] = vendas_por_sede_setor['Localidade da sede'].replace(siglas_para_estados)
display(valor_setor_localidadeT.head())
display(vendas_por_sede_setor.head())
```
