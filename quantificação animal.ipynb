```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel("Produção animal.xlsx", sheet_name = "dados")
df2 = pd.read_excel("Produção animal.xlsx", sheet_name = "fatores de transformação")
```

```python
df1 = df1.replace("-", np.nan)
display(df1)
display(df2)
```

```python
df = pd.DataFrame()
```

```python
df["Municípios"] = df1["Município"]

df["Resíduos Bovino (t)"] = (365 * df1["Bovino"] * df2.loc[0, 'Produção Esterco '] * 
                              (df2.loc[0, 'Conteudo Solido'] / 100) * 
                              (df2.loc[0, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Galinaceos (t)"] = (365 * df1["Galinaceos"] * df2.loc[1, 'Produção Esterco '] * 
                                   (df2.loc[1, 'Conteudo Solido'] / 100) * 
                                   (df2.loc[1, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Caprino (t)"] = (365 * df1["Caprino"] * df2.loc[2, 'Produção Esterco '] * 
                               (df2.loc[2, 'Conteudo Solido'] / 100) * 
                               (df2.loc[2, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Suino (t)"] = (365 * df1["Suino"] * df2.loc[3, 'Produção Esterco '] * 
                             (df2.loc[3, 'Conteudo Solido'] / 100) * 
                             (df2.loc[3, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Ovino (t)"] = (365 * df1["Ovino"] * df2.loc[4, 'Produção Esterco '] * 
                             (df2.loc[4, 'Conteudo Solido'] / 100) * 
                             (df2.loc[4, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Equino (t)"] = (365 * df1["Equino"] * df2.loc[5, 'Produção Esterco '] * 
                              (df2.loc[5, 'Conteudo Solido'] / 100) * 
                              (df2.loc[5, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Bubalino (t)"] = (365 * df1["Bubalino"] * df2.loc[6, 'Produção Esterco '] * 
                                (df2.loc[6, 'Conteudo Solido'] / 100) * 
                                (df2.loc[6, 'Fator de Coleção'] / 100)) / 1000 

df["Resíduos Codornas (t)"] = (365 * df1["Codornas"] * df2.loc[7, 'Produção Esterco '] * 
                               (df2.loc[7, 'Conteudo Solido'] / 100) * 
                               (df2.loc[7, 'Fator de Coleção'] / 100)) / 1000 

```

```python
desvio_padrao = 0.05

def monte_carlo_simulation(residuo_biomassa, tipo_residuo, n_inter=10000):
    if pd.isna(residuo_biomassa):
        return np.nan  
    
    
    rendimento_conversao = (df2.loc[df2["Tipo de animal"] == tipo_residuo, "Rendimento do Biochar "].values[0] / 100)
    desvio_padrao_rendimento = rendimento_conversao * desvio_padrao
   
    rendimento_conversao_com_variacao = np.random.normal(rendimento_conversao, desvio_padrao_rendimento)
    quantidade_biochar = residuo_biomassa * rendimento_conversao_com_variacao
    return np.mean([quantidade_biochar])


def calcular_biochar_para_coluna(df, nome_coluna, tipo_residuo):
    return df[nome_coluna].apply(lambda x: monte_carlo_simulation(x, tipo_residuo))

df["Biochar Bovino (t)"] = calcular_biochar_para_coluna(df, "Resíduos Bovino (t)", "Bovino")
df["Biochar Galinaceos (t)"] = calcular_biochar_para_coluna(df, "Resíduos Galinaceos (t)", "Galinaceos")
df["Biochar Caprino (t)"] = calcular_biochar_para_coluna(df, "Resíduos Caprino (t)", "Caprino")
df["Biochar Suino (t)"] = calcular_biochar_para_coluna(df, "Resíduos Suino (t)", "Suino")
df["Biochar Ovino (t)"] = calcular_biochar_para_coluna(df, "Resíduos Ovino (t)", "Ovino")
df["Biochar Equino (t)"] = calcular_biochar_para_coluna(df, "Resíduos Equino (t)", "Equino")
df["Biochar Bubalino (t)"] = calcular_biochar_para_coluna(df, "Resíduos Bubalino (t)", "Bubalino")
df["Biochar Codornas (t)"] = calcular_biochar_para_coluna(df, "Resíduos Codornas (t)", "Codornas")

print(df)
```
