```python
import pandas as pd
df = pd.read_excel("C:\\Users\\Aluno\\Documents\\Gabriel\\PRH 55.1\\Mapeamento - Banco de dados\\BAHIA.xlsx", sheet_name="Área colhida (Hectares)")
display(df)
```

```python
new_df = []
for culture in df.columns[1:]:
    for index, row in df.iterrows():
        value = row[culture]
        if value != '-':
            new_df.append({
                'Cidade': row['Município'],
                'Área colhida (Hectares)': value,  
                'Nome': culture
            })
newdf = pd.DataFrame(new_df)
display(newdf)
```

```python
df2 = pd.read_excel("C:\\Users\\Aluno\\Documents\\Gabriel\\PRH 55.1\\Mapeamento - Banco de dados\\BAHIA.xlsx", sheet_name="Quantidade produzida (Tonela...")
display(df2)
```

```python
new_df2 = []
for culture in df2.columns[1:]:
    for index, row in df2.iterrows():
        value = row[culture]
        if value != '-':
            new_df2.append({
                'Cidade': row['Município'],
                'Produção (toneladas)': value,  
                'Nome': culture
            })
newdf2 = pd.DataFrame(new_df2)
display(newdf2)
```

```python
with pd.ExcelWriter("C:\\Users\\Aluno\\Documents\\Gabriel\\PRH 55.1\\Mapeamento - Banco de dados\\resultados\\ resultados BAHIA.xlsx", engine='openpyxl') as writer:
    newdf.to_excel(writer, sheet_name='Área Colhida (hectares)', index=False)
    newdf2.to_excel(writer, sheet_name='Produção (toneladas)', index=False)
```
